Special exceptions, from https://github.com/ranked-vote/rcv-data-format/blob/master/README.md#normalized-vs-raw:

- UNDERVOTE: the ballot did not list a candidate for this rank.
- OVERVOTE: the ballot voted for multiple candidates in the rank, invalidating the vote for that rank (and possibly others, depending on the jurisdiction’s rules.)
- WRITE_IN: the ballot listed a write-in candidate, and the raw voting data does not indicate that candidate’s name.

```python
import pyrankvote
from pyrankvote import Candidate, Ballot

kiss = Candidate('Bob Kiss')
montroll = Candidate('Andy Montroll')
wright = Candidate('Kurt Wright')
smith = Candidate('Dan Smith')
simpson = Candidate('James Simpson')
```

```python
candidates = [kiss, montroll, wright, smith, simpson]
```

```python
df.head()
```

```python
# Getting each candidate per ballot, joining them with a comma, making into list

ballots_with_candidates_ranked = df.groupby(['ballot_id'])['choice'].transform(lambda x: ', '.join(x)).tolist()
```

```python
def remove_writeins_from_ballots(list_of_ballots, value_to_remove):
    """
    list_of_ballots: list of ballots, each ballot is a string containing the candidates in rank order.
    value_to_remove: string that a ballot contains as a condition for its deletion
    """
    indices_to_drop = []
    
    for index, value in enumerate(ballots_with_candidates_ranked):
        if value_to_remove in value:
            indices_to_drop.append(index)
            
    for index in sorted(indices_to_drop, reverse=True):
        del list_of_ballots[index]
        
    return list_of_ballots
```

```python
# Removing Write-in

ballots_with_candidates_ranked = remove_writeins_from_ballots(ballots_with_candidates_ranked, 'Write-in')
```

```python
# Removing $UNDERVOTE

ballots_with_candidates_ranked = remove_writeins_from_ballots(ballots_with_candidates_ranked, '$UNDERVOTE')
```

```python
# Removing $OVERVOTE

ballots_with_candidates_ranked = remove_writeins_from_ballots(ballots_with_candidates_ranked, '$OVERVOTE')
```

```python
def initialize_candidate_objects_in_ballots(list_of_ballots_with_candidates):
    """
    list_of_ballots_with_candidates: list of ballots, each containing a 
    string of candidate names
    
    goal: take each candidate name in each ballot & declare them as Candidate
    objects, as per the PyRankVote library
    
    return: list of ballots, each containing proper Candidate objects
    """
    candidates_array = []
    
    list_of_candidate_lists = [cand.split(",") for cand in list_of_ballots_with_candidates]
        
    for candidate_list in list_of_candidate_lists:
        ballot = []
        for candidate in candidate_list:
            candidate = candidate.strip(' ')
            ballot.append(Candidate(candidate))
        candidates_array.append(ballot)
        
    return candidates_array
```

```python
ballots_with_candidates_ranked = initialize_candidate_objects_in_ballots(ballots_with_candidates_ranked)
```

```python
from pyrankvote.models import DuplicateCandidatesError

def initialize_ballot_objects(list_of_candidate_objects):
    """
    list_of_candidate_objects: list of lists. Each list containing, in rank order, the candidates
    per ballot. Each candidate is an initialized Candidate object.

    """
    ballots = []
    
    for candidate_group in range(len(list_of_candidate_objects)):
        try:
            ballot = Ballot(ranked_candidates=ballots_with_candidates_ranked[candidate_group])
            ballots.append(ballot)
        except:
            pass
    
    return ballots
```

```python
ballots = initialize_ballot_objects(ballots_with_candidates_ranked)
```

```python
election_result = pyrankvote.instant_runoff_voting(candidates, ballots)
```
